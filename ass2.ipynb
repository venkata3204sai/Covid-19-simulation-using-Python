```python
#sample_population = pd.DataFrame(columns=["person_id", "age_group_name", "country"])
records = []
person_id = 0
for i in ["Sweden","Australia"]:
    total_population = df.loc[df["country"] == i, "population"].item()
    total_sample = int(total_population / 1e6)
    for j in ["less_5", "5_to_14", "15_to_24","25_to_64", "over_65"]:
        percent_of_population = df.loc[df["country"] == i, j].item()
        number_of_sample = total_sample * percent_of_population / 100
        for _ in range(int(number_of_sample)):
            person_id+=1
            records.append([person_id,j,i])
#new_persons = pd.DataFrame(records, columns=sample_population.columns)
#sample_population = pd.concat([sample_population, new_persons], ignore_index=True)

sample_population = pd.DataFrame(records, columns=["person_id", "age_group_name", "country"])
```

```python
sample_population
```

```python
transition_probs = {
    'less_5':
{
    'H':{'H':0.9, 'I':0.1}, 
    'I':{'I':0, 'S':0.5, 'M':0.5}, 
    'S':{'S':0, 'D':0.01, 'M':0.99}, 
    'D':{'D':1}, 
    'M':{'M':0, 'H':1}
},
    '5_to_14':
{
    'H':{'H':0.9, 'I':0.1}, 
    'I':{'I':0, 'S':0.5, 'M':0.5}, 
    'S':{'S':0, 'D':0.01, 'M':0.99}, 
    'D':{'D':1}, 
    'M':{'M':0, 'H':1}
},
    '15_to_24':
{
    'H':{'H':0.7, 'I':0.3}, 
    'I':{'I':0, 'S':0.5, 'M':0.5}, 
    'S':{'S':0, 'D':0.05, 'M':0.95}, 
    'D':{'D':1}, 
    'M':{'M':0, 'H':1}
},
    '25_to_64':
{
    'H':{'H':0.7, 'I':0.3}, 
    'I':{'I':0, 'S':0.5, 'M':0.5}, 
    'S':{'S':0, 'D':0.2, 'M':0.8}, 
    'D':{'D':1}, 
    'M':{'M':0, 'H':1}
},
    'over_65':
{
    'H':{'H':0.7, 'I':0.3}, 
    'I':{'I':0, 'S':0.5, 'M':0.5}, 
    'S':{'S':0, 'D':0.4, 'M':0.6}, 
    'D':{'D':1}, 
    'M':{'M':0, 'H':1}
}
}

holding_times = {
    'less_5':

    {'H':0, 'I':4, 'S':14, 'D':0, 'M':120}

,
    '5_to_14':

{'H':0, 'I':4, 'S':14, 'D':0, 'M':120}
,
    '15_to_24':

{'H':0, 'I':4, 'S':14, 'D':0, 'M':120}
,
    '25_to_64':

{'H':0, 'I':4, 'S':14, 'D':0, 'M':120}
,
    'over_65':

{'H':0, 'I':4, 'S':14, 'D':0, 'M':120}


}
```

```python
start_date = '2021-04-01'
end_date = '2022-04-30'
dates = pd.date_range(start = start_date, end = end_date)
records = []
for _, row in sample_population.iterrows():
    state = "H"
    staying_days = 0
    previous_state = "H"
    for date in dates:
        records.append([
            row["person_id"],
            row["age_group_name"],
            row["country"],
            date,
            state,
            staying_days,
            previous_state,
        ])
        staying_days += 1
        if staying_days >= holding_times[row["age_group_name"]][state]:
            previous_state = state
            state = np.random.choice(
                list(transition_probs[row["age_group_name"]][state].keys()),
                p = list(transition_probs[row["age_group_name"]][state].values())
            )
            staying_days = 0
timeseries_data = pd.DataFrame(records, columns= list(sample_population.columns) + ["date", "state", "staying_days", "previous_state"])
```

```python
timeseries_data.head(500)
```

```python
timeseries_data = timeseries_data.groupby(["date", "country", "state"]).size().reset_index(name="count")
```

```python
summarised_data = timeseries_data.pivot_table(
    index=["date", "country"],
    columns="state",
    values="count",
    fill_value=0
).reset_index()
for col in ["D", "H", "I", "M", "S"]:
    if col not in summarised_data.columns:
        summarised_data[col] = 0
summarised_data.to_csv("xyz.csv", index=False)
summarised_data
```
